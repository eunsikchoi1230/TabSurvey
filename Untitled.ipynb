```python
import shap
import pickle

with open('data/split_data.pkl', 'rb') as f:
    X_train, y_train, X_test, y_test = pickle.load(f)

with open('output/CMCscreening/None/CatBoost/models/m_.pkl', 'rb') as f:
    model = pickle.load(f)
```

```python
importances = model.feature_importances_

columns = 'sex,age,height,weight,BMI,waist,muscle_mass,bodyfat_mass,bodyfat_percent,WHR,FEV1,PR,Hb,Prot_Test,chol,TG,HDL,AST_SGOT,ALT_SGPT,CAD_echo,PAOD_sono,CAD_CT,stroke_MR,PAOD_dx,IFG,dyslipid,glaucoma,cancer_dx,alcohol,smoke,marriage,education,job,income,menopause,chest_pain,CAD_fx,IPSS,exercise_intensity,glucose,HbA1c,HTN_class,DM_class,CAD_class,Dyslipidemia_class'
columns = columns.split(',')[:-4]
```

```python
column_importances = list(zip(columns, importances))
sorted_importances = sorted(column_importances, key=lambda x: x[1], reverse=True)
sorted_columns, sorted_importances_values = zip(*sorted_importances)
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

fig, ax = plt.subplots(figsize=(7.5, 9), dpi=600)
ax.barh(sorted_columns[:20], sorted_importances_values[:20], color='royalblue')
ax.set_xlabel('Feature Importance', fontsize=15)
# ax.set_title('Feature Importances from CatBoost Model')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Adding minor ticks
ax.xaxis.set_minor_locator(ticker.AutoMinorLocator())

# Invert y axis to have the most important feature at the top
plt.gca().invert_yaxis()

# Set font size for tick labels
ax.tick_params(axis='both', which='major', labelsize=10)
ax.tick_params(axis='x', which='minor', labelsize=8)  # Minor tick label size
plt.savefig('output/feature_importance.png', dpi=600)
plt.show()
```
